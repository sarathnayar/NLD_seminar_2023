```python
def plot_nullclines(a=3, b=0.2, I=0, eps=0):
    # u nullcline
    u = np.linspace(-1, 2, 100)
    w = a*u*(u-b)*(1-u)+I
    plt.plot(u, w)
    # w nullcline
    w = np.linspace(-1, 1)
    u = w
    plt.plot(u, w)
    plt.xlim(-1, 2)
    plt.ylim(-1, 1)
    plt.xlabel('u')
    plt.ylabel('w')
    
plot_nullclines()
```
